```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

```python
df.index.name
```

```python
df.index.name = 'job_index'
```

```python
df
```

```python
df.index.dtype
```

```python
df_egypt = df[df['job_country'] == 'Egypt']
```

```python
df_egypt
```

```python
df_egypt.reset_index(inplace=True)
```

```python
df_egypt
```

```python
df_egypt.set_index('job_index' , inplace=True)
```

```python
df_egypt
```

```python
median_pivot = df_egypt.pivot_table(
    values='salary_year_avg',
    index='job_title_short',
    aggfunc=['median','min','max']
)
```

```python
median_pivot
```

```python
#the by is tuple because we have two levels of naming
median_pivot.sort_values(by=('median' , 'salary_year_avg') , ascending=False)
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
medain_salary_year = df['salary_year_avg'].median()
```

```python
medain_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(medain_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(medain_salary_hour)
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

#the number of droped columns
print('length of original df :', len(df_filled))
print('length of drop duplicate df :', len(df_unique))
print('Number of droped rows :', len(df_filled) - len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

#the number of droped columns
print('length of original df :', len(df_filled))
print('length of drop duplicate df :', len(df_unique))
print('Number of droped rows :', len(df_filled) - len(df_unique))

```
